 Mengimpor dan memeriksa dataset "Student Performance" untuk melihat struktur data dan informasi yang terkandung di dalamnya.

```python
import pandas as pd

# Load the dataset
data_path = 'StudentsPerformance.csv'
students_data = pd.read_csv(data_path)

# Display the first few rows of the dataset and its summary
students_data.head(), students_data.describe(), students_data.info()
```

Visualisasi Distribusi Skor
Berikut adalah visualisasi distribusi skor matematika, membaca, dan menulis:

Skor Matematika: Distribusi skor matematika terlihat normal dengan beberapa siswa mendapatkan skor sangat rendah, mengindikasikan ada kesulitan dalam subjek ini.
Skor Membaca: Distribusi skor membaca lebih merata dengan kecenderungan sedikit lebih tinggi daripada matematika, menunjukkan kemampuan membaca yang lebih baik secara umum.
Skor Menulis: Skor menulis juga menunjukkan distribusi yang mirip dengan membaca, namun dengan sedikit ekor yang lebih panjang ke arah skor yang lebih rendah, menandakan adanya siswa yang mengalami kesulitan lebih dalam menulis.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotting the distribution of scores
sns.histplot(students_data['math score'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Math Scores')

sns.histplot(students_data['reading score'], bins=20, kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Distribution of Reading Scores')

sns.histplot(students_data['writing score'], bins=20, kde=True, ax=axes[2], color='salmon')
axes[2].set_title('Distribution of Writing Scores')

# Display the plots
plt.tight_layout()
plt.show()
```

Visualisasi Skor Ujian Berdasarkan Gender
Berikut adalah temuan dari visualisasi skor ujian berdasarkan gender:

Skor Matematika: Siswa laki-laki cenderung mendapatkan skor yang lebih tinggi dalam matematika dibandingkan dengan siswi perempuan. Distribusi skor laki-laki lebih terpusat pada nilai yang lebih tinggi.
Skor Membaca dan Menulis: Siswa perempuan cenderung mendominasi dalam skor membaca dan menulis, dengan median dan rentang skor yang lebih tinggi daripada laki-laki. Ini menunjukkan bahwa perempuan mungkin memiliki kemampuan literasi yang lebih kuat.

```python
# Create a figure with subplots for scores by gender
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotting the scores by gender
sns.boxplot(x='gender', y='math score', data=students_data, ax=axes[0], palette='cool')
axes[0].set_title('Math Scores by Gender')

sns.boxplot(x='gender', y='reading score', data=students_data, ax=axes[1], palette='cool')
axes[1].set_title('Reading Scores by Gender')

sns.boxplot(x='gender', y='writing score', data=students_data, ax=axes[2], palette='cool')
axes[2].set_title('Writing Scores by Gender')

# Display the plots
plt.tight_layout()
plt.show()
```

Visualisasi Skor Ujian Berdasarkan Tingkat Pendidikan Orang Tua
Temuan dari visualisasi skor ujian berdasarkan tingkat pendidikan orang tua adalah sebagai berikut:

Skor Matematika: Terlihat adanya tren meningkat di skor matematika dengan meningkatnya tingkat pendidikan orang tua, khususnya di antara orang tua yang memiliki gelar 'master's degree'.
Skor Membaca dan Menulis: Sama seperti matematika, skor membaca dan menulis juga cenderung lebih tinggi bagi siswa yang orang tuanya memiliki pendidikan yang lebih tinggi. Hal ini menunjukkan potensi pengaruh positif dari lingkungan pendidikan keluarga terhadap kemampuan akademis siswa.

```python
# Create a figure with subplots for scores by parental level of education
fig, axes = plt.subplots(3, 1, figsize=(10, 18))

# Plotting the scores by parental level of education
sns.boxplot(x='math score', y='parental level of education', data=students_data, ax=axes[0], palette='pastel')
axes[0].set_title('Math Scores by Parental Level of Education')

sns.boxplot(x='reading score', y='parental level of education', data=students_data, ax=axes[1], palette='pastel')
axes[1].set_title('Reading Scores by Parental Level of Education')

sns.boxplot(x='writing score', y='parental level of education', data=students_data, ax=axes[2], palette='pastel')
axes[2].set_title('Writing Scores by Parental Level of Education')

# Display the plots
plt.tight_layout()
plt.show()
```

Visualisasi Skor Ujian Berdasarkan Kursus Persiapan Tes
Dari visualisasi ini, kita mendapatkan insight sebagai berikut:

Skor Matematika, Membaca, dan Menulis: Siswa yang telah menyelesaikan kursus persiapan tes cenderung memiliki skor yang lebih tinggi di semua bidang ujian dibandingkan dengan mereka yang tidak mengikuti kursus. Hal ini menunjukkan bahwa kursus persiapan tes bisa sangat membantu dalam meningkatkan kinerja akademik siswa.

```python
# Create a figure with subplots for scores by test preparation course
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plotting the scores by test preparation course
sns.boxplot(x='test preparation course', y='math score', data=students_data, ax=axes[0], palette='autumn')
axes[0].set_title('Math Scores by Test Preparation Course')

sns.boxplot(x='test preparation course', y='reading score', data=students_data, ax=axes[1], palette='autumn')
axes[1].set_title('Reading Scores by Test Preparation Course')

sns.boxplot(x='test preparation course', y='writing score', data=students_data, ax=axes[2], palette='autumn')
axes[2].set_title('Writing Scores by Test Preparation Course')

# Display the plots
plt.tight_layout()
plt.show()
```
