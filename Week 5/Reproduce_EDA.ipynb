Memuat data dan melihat beberapa baris pertama untuk memahami struktur datanya

```python
import pandas as pd

# Load the data from the uploaded file
data_path = 'data.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset
data.head()
```

Statistik deskriptif memberikan gambaran singkat mengenai nilai numerik dalam dataset:

Year: Data meliputi mobil yang dibuat dari tahun 1990 hingga 2017.
Engine HP: Rata-rata tenaga kuda (HP) adalah 249, dengan mobil terendah memiliki 55 HP dan tertinggi 1001 HP.
Engine Cylinders: Rata-rata jumlah silinder adalah sekitar 5.6, dengan minimal 0 dan maksimal 16.
Number of Doors: Sebagian besar mobil memiliki 2 hingga 4 pintu.
MPG (Miles Per Gallon): MPG di jalan raya dan dalam kota berkisar dari rendah hingga sangat tinggi, menunjukkan variasi yang luas dalam efisiensi bahan bakar.
Popularity: Popularitas merek bervariasi secara signifikan.
MSRP (Manufacturer's Suggested Retail Price): Harga jual yang disarankan oleh produsen berkisar dari $2,000 hingga lebih dari $2 juta, menunjukkan variasi besar dalam harga.

```python
# Get the statistical summary of the numerical columns
data.describe()
```

Dataset memiliki beberapa nilai yang hilang di beberapa kolom:

Engine Fuel Type: 3 nilai hilang
Engine HP: 69 nilai hilang
Engine Cylinders: 30 nilai hilang
Number of Doors: 6 nilai hilang
Market Category: 3742 nilai hilang (jumlah yang cukup besar)

```python
# Check for missing values in the dataset
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

Semua nilai yang hilang untuk Engine HP, Engine Cylinders, Number of Doors, dan Engine Fuel Type telah diisi dengan median atau modus masing-masing kolom. Untuk Market Category, nilai yang hilang masih ada karena pertimbangan sebelumnya

```python
# Fill missing numerical values with the median of their respective columns
for column in ['Engine HP', 'Engine Cylinders', 'Number of Doors']:
    median_value = data[column].median()
    data[column].fillna(median_value, inplace=True)

# Since 'Engine Fuel Type' has only 3 missing values, we'll fill them with the mode
mode_value = data['Engine Fuel Type'].mode()[0]
data['Engine Fuel Type'].fillna(mode_value, inplace=True)

# Recheck for missing values
data.isnull().sum()
```

Mean Squared Error (MSE): 1,067,504,769
Root Mean Squared Error (RMSE): 32,672.69
R-squared: 0.524

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

# Prepare the features and target variable
features = data[['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG']]
target = data['MSRP']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=0)

# Train a linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = linear_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

mse, rmse, r2
```

R-squared menunjukkan bahwa sekitar 52.4% variasi dalam harga MSRP dapat dijelaskan oleh fitur yang kita gunakan dalam model.

```python
# Apply polynomial transformation to the features
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly_train = poly_features.fit_transform(X_train)
X_poly_test = poly_features.transform(X_test)

# Train a linear regression model on the transformed features
poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

# Make predictions and evaluate the model
y_poly_pred = poly_model.predict(X_poly_test)
poly_mse = mean_squared_error(y_test, y_poly_pred)
poly_rmse = np.sqrt(poly_mse)
poly_r2 = r2_score(y_test, y_poly_pred)

poly_mse, poly_rmse, poly_r2
```

Dari hasil ini, kita melihat peningkatan signifikan dalam kinerja model dengan penggunaan transformasi polinomial. R-squared meningkat menjadi sekitar 68.6%, yang menunjukkan bahwa model yang lebih kompleks dengan basis fungsi lebih baik dalam menjelaskan variasi harga MSRP dibandingkan dengan model regresi linear biasa.
